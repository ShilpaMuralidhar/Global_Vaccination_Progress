```python
import pandas as pd
import numpy as np
```

```python
vax_name = []
vax_country = []
vax_year = []
vax_pct= []
vax_countryyear = []
vax_vaxcountryyear = []

vaccine_list = ["BCG","DTP1","DTP3","HEPB3","HEPBB","HIB","IPV1","MCV1","MCV2","PCV3","POL3","RCV1","ROTAC","YFV"]
# vaccine_list = ["BCG", "DTP1","DTP3", "HEPB3","HEPBB","HIB","IPV1","MCV1","MCV2","PCV3"]

for vax in vaccine_list:
    vax_csv = "../csv/14vaccine_data/2018_"+ vax+".csv"
    vax_data = pd.read_csv(vax_csv)
    vax_df = pd.DataFrame(vax_data)
    vax_df = vax_df.dropna(thresh= 9) 
    col_names = vax_data.columns
    for i in range(4, len(col_names)):
        for j in range(vax_df["country"].count()):
            vax_name.append(vax)
            vax_country.append(vax_df.iloc[j, 2])
            vax_year.append(col_names[i])
            vax_pct. append(vax_df.iloc[j,i])
            vax_countryyear.append(col_names[i]+vax_df.iloc[j,2])
            vax_vaxcountryyear.append(col_names[i]+vax+vax_df.iloc[j,2])

all_vax = pd.DataFrame(list(zip(vax_name, vax_country, vax_year, vax_countryyear, vax_vaxcountryyear, vax_pct)), columns = ["Vaccine_Name","Country","Year","Life_Key","Vaccine_Key","Coverage"])
all_vax.to_csv("../csv/clean_data/vaccine_stack.csv")
all_vax
```
