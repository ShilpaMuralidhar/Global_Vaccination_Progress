```python
import pandas as pd
import numpy as np
import cufflinks as cf
import plotly.offline
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
```

```python
dtp1 = "/Users/shilpa/Desktop/Desktop/RICE_DATA/Global_Vaccination_Progress/Shilpa/Plots/Clean_Plots/DTP1/df_DTP1_All_countries copy.csv"
DTP1 = pd.read_csv(dtp1)
DTP1.head(5)
reversed_df_DTP1= DTP1.sort_values(['year'], ascending= True)
reversed_df_DTP1
```

```python
from bubbly.bubbly import bubbleplot 

figure = bubbleplot(dataset=reversed_df_DTP1, x_column='Target', y_column='Coverage', 
    bubble_column='country', time_column='year', size_column='values', color_column='Subgroup',
    x_title="Target", y_title="% covered", title='DTP1 vaccination',
    x_logscale=True, scale_bubble=3, height=650)

plotly.offline.iplot(figure, config={'scrollzoom': True})
```
