```python
import pandas as pd
import numpy as np
import cufflinks as cf
import plotly.offline
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
```

```python
ipv1 = "df_IPV1_All_countries.csv"

df_ipv1 = pd.read_csv(ipv1)
df_ipv1.head(5)
```

```python
from bubbly.bubbly import bubbleplot 

figure = bubbleplot(dataset=df_ipv1, x_column='Target', y_column='Coverage', 
    bubble_column='country', time_column='year', size_column='values', color_column='Subgroup',
    x_title="Target", y_title="% covered", title='IPV1 vaccination',
    x_logscale=True, scale_bubble=3, height=650)

plotly.offline.iplot(figure, config={'scrollzoom': True})
```
